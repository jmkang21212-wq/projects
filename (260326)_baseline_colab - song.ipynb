```python
# import os, re, math, random
# import pandas as pd
# from PIL import Image
# from torch.utils.data import Dataset, DataLoader
# from dataclasses import dataclass
# import torch
# from typing import Dict, List, Any
# from transformers import (
#     AutoModelForVision2Seq,
#     AutoProcessor,
#     BitsAndBytesConfig,
#     get_linear_schedule_with_warmup
# )
# from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
# from tqdm import tqdm

# # 이미지 로드 시 픽셀 제한 해제
# Image.MAX_IMAGE_PIXELS = None

# # 디바이스 GPU 우선 사용 설정
# device = "cuda" if torch.cuda.is_available() else "cpu"
# print("Device:", device)

# # 사전 학습 모델 정의
# MODEL_ID = "Qwen/Qwen2.5-VL-32B-Instruct"
# IMAGE_SIZE = 384
# MAX_NEW_TOKENS = 8
# SEED = 42
# random.seed(SEED); torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)

# # 데이터셋 로드
# train_df = pd.read_csv("/content/train.csv")
# test_df  = pd.read_csv("/content/test.csv")

# # 학습데이터 200개만 추출
# train_df = train_df.sample(n=200, random_state=SEED).reset_index(drop=True)
```

```python
import os, re, math, random
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from dataclasses import dataclass
import torch
from typing import Dict, List, Any
from transformers import (
    Qwen2_5_VLForConditionalGeneration,
    AutoProcessor,
    BitsAndBytesConfig,
    get_linear_schedule_with_warmup
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from tqdm import tqdm

Image.MAX_IMAGE_PIXELS = None
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# ✅ 32B — A100 80GB에서 4bit 양자화로 안정적으로 동작
MODEL_ID = "Qwen/Qwen2.5-VL-32B-Instruct"
MAX_NEW_TOKENS = 8
SEED = 42
random.seed(SEED); torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)

# ✅ 4bit 양자화 설정 — A100이라도 32B는 필수
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,  # A100은 bfloat16 지원
    bnb_4bit_use_double_quant=True,          # 메모리 추가 절약
)

# ✅ 모델 로드
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
)

# ✅ Processor — dynamic resolution 제한으로 VRAM 안정화
processor = AutoProcessor.from_pretrained(
    MODEL_ID,
    min_pixels=256 * 28 * 28,
    max_pixels=512 * 28 * 28,  # 너무 크면 32B + 이미지 토큰이 VRAM 폭발
)

# ✅ LoRA 설정 — 32B에 맞게 rank 상향
lora_config = LoraConfig(
    r=16,                        # 32B면 16 권장
    lora_alpha=32,
    target_modules="all-linear",
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)

model = prepare_model_for_kbit_training(model)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# 데이터셋 로드
train_df = pd.read_csv("/content/train.csv")
test_df  = pd.read_csv("/content/test.csv")
train_df = train_df.sample(n=200, random_state=SEED).reset_index(drop=True)
```

```python
# ✅ 양자화 — bfloat16으로 변경 (A100은 bfloat16이 float16보다 안정적)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,  # float16 → bfloat16
)

# ✅ 프로세서 — dynamic resolution 방식으로 변경
processor = AutoProcessor.from_pretrained(
    MODEL_ID,
    min_pixels=256 * 28 * 28,   # Qwen 권장 방식
    max_pixels=512 * 28 * 28,   # IMAGE_SIZE 고정 방식보다 안정적
    trust_remote_code=True,
)

# ✅ 모델 로드 — torch_dtype 명시 추가
base_model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,  # 추가
    trust_remote_code=True,
)

# 양자화 모델 준비
base_model = prepare_model_for_kbit_training(base_model)
base_model.gradient_checkpointing_enable()

# ✅ LoRA — 32B 모델에 맞게 r/alpha 상향
lora_config = LoraConfig(
    r=16,           # 8 → 16 (32B 모델 표현력 활용)
    lora_alpha=32,  # 16 → 32 (alpha는 보통 r의 2배)
    lora_dropout=0.05,
    bias="none",
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    task_type="CAUSAL_LM",
)

# PEFT 모델 생성
model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters()
```

# 프롬프트 템플릿

#### 실습 참고 내용

    챕터 5-1 PEFT(파라미터 효율적 튜닝)
    - 프롬프트 템플릿 : convert_to_chatml(), formatting_prompts_func()

```python
# System prompt — 판단 기준 명시
SYSTEM_INSTRUCT = (
    "You are an expert recycling classification assistant. "
    "You will be shown an image and a multiple choice question about recycling. "
    "Analyze the image carefully before answering. "
    "Output only a single lowercase letter: a, b, c, or d. "
    "No explanation, no punctuation, no extra text."
)

# 프롬프트 — 이미지 먼저 보도록 유도 + 한국어 통일
def build_mc_prompt(question, a, b, c, d):
    return (
        f"다음 이미지를 자세히 관찰하고 질문에 답하세요.\n\n"
        f"{question}\n"
        f"(a) {a}\n"
        f"(b) {b}\n"
        f"(c) {c}\n"
        f"(d) {d}\n\n"
        f"반드시 a, b, c, d 중 하나의 소문자 한 글자만 출력하세요."
    )
```

# Custom Dataset, Collator

#### 실습 참고 내용

    챕터 1-2 MLP 구현
    - TensorDataset()

    챕터 5-2 데이터 생성 및 파인튜닝 (향후 학습 분량)
    - IntentDataset()

```python
# 커스텀 데이터셋
class VQAMCDataset(Dataset):
    def __init__(self, df, processor, train=True):
        self.df = df.reset_index(drop=True)
        self.processor = processor
        self.train = train

    def __len__(self): return len(self.df)

    def __getitem__(self, i):
        row = self.df.iloc[i]
        img = Image.open(row["path"]).convert("RGB")

        q = str(row["question"])
        a, b, c, d = str(row["a"]), str(row["b"]), str(row["c"]), str(row["d"])
        user_text = build_mc_prompt(q, a, b, c, d)

        messages = [
            {"role":"system","content":[{"type":"text","text":SYSTEM_INSTRUCT}]},
            {"role":"user","content":[
                {"type":"image","image":img},
                {"type":"text","text":user_text}
            ]}
        ]
        if self.train:
            gold = str(row["answer"]).strip().lower()
            messages.append({"role":"assistant","content":[{"type":"text","text":gold}]})

        return {"messages": messages, "image": img}

# 데이터 콜레이터
@dataclass
class DataCollator:
    processor: Any
    train: bool = True

    def __call__(self, batch):
        texts, images = [], []
        for sample in batch:
            messages = sample["messages"]
            img = sample["image"]

            text = self.processor.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=False
            )
            texts.append(text)
            images.append(img)

        enc = self.processor(
            text=texts,
            images=images,
            padding=True,
            return_tensors="pt"
        )

        if self.train:
            enc["labels"] = enc["input_ids"].clone()

        return enc
```

# DataLoader

#### 실습 참고 내용

    챕터 3-1 Transfer Learning 기반의 CNN 모델 학습
    - 데이터로더 정의 : DataLoader()

```python
# 검증용 데이터 분리
split = int(len(train_df)*0.9)
train_subset, valid_subset = train_df.iloc[:split], train_df.iloc[split:]

# VQAMCDataset 형태로 변환
train_ds = VQAMCDataset(train_subset, processor, train=True)
valid_ds = VQAMCDataset(valid_subset, processor, train=True)

# 데이터로더
train_loader = DataLoader(train_ds, batch_size=1, shuffle=True, collate_fn=DataCollator(processor, True), num_workers=0)
valid_loader = DataLoader(valid_ds, batch_size=1, shuffle=False, collate_fn=DataCollator(processor, True), num_workers=0)
```

# fine-tuning

- 200개만 학습 : 10~20분 소요

#### 실습 참고 내용

    챕터 1-2 MLP 구현
    - 모델 정의 : SimpleMLP(), SequentialMLP()

    챕터 3-1 Transfer Learning 기반의 CNN 모델 학습
    - 학습 루프 : 문제 6: 모델 학습을 위한 반복문
    - 추론 : with torch.no_grad(), model.eval()

```python
from tqdm.auto import tqdm

model = model.to(device)

# ✅ 배치 누적 증가 — 200개 소규모 데이터에서 안정적인 학습
GRAD_ACCUM = 8          # 4 → 8 (유효 배치 크기 2배)
NUM_EPOCHS = 5          # 1 → 5 (소규모 데이터는 에폭을 늘려야 함)
MAX_GRAD_NORM = 1.0     # gradient clipping 추가

# ✅ 옵티마이저 — weight_decay 추가로 과적합 방지
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=2e-4,            # 1e-4 → 2e-4 (LoRA fine-tuning 권장값)
    weight_decay=0.01,  # 추가
    betas=(0.9, 0.999),
)

num_training_steps = NUM_EPOCHS * math.ceil(len(train_loader) / GRAD_ACCUM)
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(num_training_steps * 0.1),  # 0.03 → 0.1 (워밍업 충분히)
    num_training_steps=num_training_steps,
)

# ✅ bfloat16이면 GradScaler 불필요 — A100은 bfloat16 native 지원
# scaler 제거하고 autocast만 사용
best_val_loss = float('inf')
global_step = 0

for epoch in range(NUM_EPOCHS):
    # === Train ===
    model.train()
    running = 0.0
    optimizer.zero_grad(set_to_none=True)
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS} [train]", unit="batch")

    for step, batch in enumerate(progress_bar, start=1):
        batch = {k: v.to(device) for k, v in batch.items()}

        with torch.cuda.amp.autocast(dtype=torch.bfloat16):
            outputs = model(**batch)
            loss = outputs.loss / GRAD_ACCUM

        loss.backward()         # ✅ scaler 제거
        running += loss.item()

        if step % GRAD_ACCUM == 0:
            # ✅ Gradient clipping — 학습 안정성
            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)

            optimizer.step()
            optimizer.zero_grad(set_to_none=True)
            scheduler.step()
            global_step += 1

            avg_loss = running / GRAD_ACCUM
            progress_bar.set_postfix({
                "loss": f"{avg_loss:.3f}",
                "lr": f"{scheduler.get_last_lr()[0]:.2e}",  # ✅ lr 모니터링 추가
            })
            running = 0.0

    # === Validation ===
    model.eval()
    val_loss = 0.0
    val_steps = 0

    with torch.no_grad(), torch.cuda.amp.autocast(dtype=torch.bfloat16):
        for vb in tqdm(valid_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS} [valid]", unit="batch"):
            vb = {k: v.to(device) for k, v in vb.items()}
            val_loss += model(**vb).loss.item()
            val_steps += 1

    avg_val_loss = val_loss / val_steps
    print(f"[Epoch {epoch+1}] valid loss: {avg_val_loss:.4f}")

    # ✅ Best 모델만 저장 — 마지막 에폭이 최선이 아닐 수 있음
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        SAVE_DIR = "/content/qwen2_5_vl_32b_lora_best"  # ✅ 32B로 이름 수정
        model.save_pretrained(SAVE_DIR)
        processor.save_pretrained(SAVE_DIR)
        print(f"  ✅ Best model saved (val_loss: {best_val_loss:.4f})")
```

# inference

30분~1시간 소요

#### 실습 참고 내용

    챕터4-1 RAG 기반 Customer Service AI 에이전트 개발
    - 데이터 파서 : langchain_core.output_parsers(), StrOutputParser()

    챕터 3-1 Transfer Learning 기반의 CNN 모델 학습
    - 학습 루프 : 문제 6: 모델 학습을 위한 반복문
    - 추론 : with torch.no_grad(), model.eval()

```python
from tqdm.auto import tqdm

model.eval()
preds = []

# 추론 루프
for i in tqdm(range(len(test_df)), desc="Inference", unit="sample"):
    row = test_df.iloc[i]
    img = Image.open(row["path"]).convert("RGB")
    user_text = build_mc_prompt(row["question"], row["a"], row["b"], row["c"], row["d"])

    messages = [
        {"role": "system", "content": [{"type": "text", "text": SYSTEM_INSTRUCT}]},
        {"role": "user",   "content": [
            {"type": "image", "image": img},
            {"type": "text",  "text": user_text}
        ]}
    ]

    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], images=[img], return_tensors="pt").to(device)

    with torch.no_grad():
        out_ids = model.generate(
            **inputs,
            max_new_tokens=2,
            do_sample=False,
            temperature=None,       # ✅ do_sample=False일 때 명시
            top_p=None,             # ✅ do_sample=False일 때 명시
            eos_token_id=processor.tokenizer.eos_token_id,
        )

    # ✅ 생성된 토큰만 슬라이싱 — 입력 프롬프트 제외
    generated_ids = out_ids[:, inputs["input_ids"].shape[1]:]
    output_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

    preds.append(extract_choice(output_text))

# ✅ 개선된 extract_choice — re 기반 + fallback 개선
def extract_choice(text: str) -> str:
    text = text.strip().lower()

    # 1순위: 단독 a/b/c/d 단어 매칭
    match = re.search(r'\b([abcd])\b', text)
    if match:
        return match.group(1)

    # 2순위: 붙어있는 문자라도 추출
    for ch in text:
        if ch in 'abcd':
            return ch

    # 3순위: 가장 많이 등장한 선지로 fallback (균등 분포 가정)
    from collections import Counter
    counts = Counter(c for c in text if c in 'abcd')
    if counts:
        return counts.most_common(1)[0][0]

    return "a"  # 최후 fallback

# 제출 파일 생성
submission = pd.DataFrame({"id": test_df["id"], "answer": preds})
submission.to_csv("/content/submission.csv", index=False)
print("Saved /content/submission.csv")

# ✅ 예측 분포 확인 — 특정 클래스 쏠림 체크
print("예측 분포:", submission["answer"].value_counts().sort_index())
```
